### Beautiful 解析：
   - python独有。简单便捷和高效
   - 环境安装：
      - 需要将pip源设置为国内源（阿里源、豆瓣源、网易源等)
      - windows:
         - 打开文件资源管理器（文件夹地址栏中）
         - 地址栏上面输入：%appdata%
         - 在这里面新建一个文件夹 pip
         - 在pip文件夹里面新建一个文件叫做 pip.ini，内容写下如下：
            - [global]
            - timeout = 6000
            - index-url = https://mirrors.aliyun.com/pypi/simple/
            - trusted-host = mirrors.aliyun.com
      - linux:
         - cd ～
         - mkdir ～/.pip
         - vi ～/.pip/pip.conf
         - 编辑内容，和windows 一样
      - 需要安装： pip install bs4
          bs4 在使用时候需要一个第三方库，把这个库也安装一下
          pip install lxml
   - 代码流程：
      - 核心思想： 将html文档转成Beautiful对象，然后掉用改对象中的属性和方法进行html文档制定内容的定位查找
      - 导包： from bs4 import BeautifulSoup
      - 创建 Beautiful对象：
         - 如果html 文档是来源于本地：
            - Beautiful ('open("本地的html文件")','lxml')
         - 如果 html 是来源网络：
            - Beautiful ('网络请求到的页面数据’，'lxml')
   - 属性和方法：
      - 1.根据标签名查找
         - soup.a  智能找到第一个符合要求的标签
      - 2.获取属性：
         - soup.a.attrs   获取a 所有的属性和属性值，返回一个字典
         - soup.a.attrs['href']   获取href书书属性
         - soup.a['href']   也可以简写为这种模式
      - 3.获取内容：
         - soup.a.string
         - soup.a.text
         - soup.a.get_text()
         【注意】如果标签还有标签，那么string获取到的结果为None,而其他两个可以获取文本内容
      - 4.find：找到第一个符合要求的标签
         - soup.find('a')   找到第一个符合要求的标签
         - soup.find('a', title="xxx") 
         - soup.find('a', alt="xxx")
         - soup.find('a', class="xxx")
         - soup.find('a', id="xxx")
         
      - 5.find_all: 找到所有符合要求的标签
         - soup.find_all('a')
         - soup.find_all(['a','b'])  找到所有的a和 b 标签
         - soup.find_all('a', limit=2)  限制前两个
         
      - 6.根据选择器选择制定的内容：
         - select:soup.select('#fing')
         - 常见的选择器：
            - 标签选择器 (a)
            - 类选择器 (.)
            - id 选择器 (#)
            - 层级选择器：
               div .dudu #lala .meme .xixi  下面好多级
               div > p > a > .lala       只能下面一级
            - 【注意】select 选择器返回永远是列表，需要通过下标提取指定的对象

#### 需求：起点中文网 - 斗罗大陆小说爬取
   - url：https://book.qidian.com/info/1115277#Catalog
   
   

```python
import requests
from bs4 import BeautifulSoup

# 目标url
url = 'https://book.qidian.com/info/1115277'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36"
}
def get_content(url):
    page_text = requests.get(url=url, headers=headers).text
    # 数据解析
    soup = BeautifulSoup(page_text,'lxml')
#     title = soup.select(".main-text-wrap > .text-head > h3")[0].text
    content = soup.select(".main-text-wrap > .read-content")[0].text
    return content
    
# 获取页面数据
page_text = requests.get(url=url,headers=headers).text
# 数据解析
soup = BeautifulSoup(page_text, 'lxml')
f = open("斗罗大陆-(1-4集).txt", 'a', encoding='utf-8')
for i in range(1,5):
    volume_list1 = soup.find_all('div', class_='volume')[i]
    # print(list1)
    title_url_list = volume_list1.select('ul > li > a')
    # print(title_url_list)
    for item in title_url_list:
        url = 'https:' + item['href']
        title = item.text
        content =get_content(url)
        f.write(title)
        f.write(content+"\n")
        print(title)
f.close()
print("执行完毕！")
    
```
