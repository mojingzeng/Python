用selenium爬取斗鱼平台的所有房间的 房间名，主播名字，主播类别，房间图片，房间url, 人气值


```python
from selenium import webdriver
import time

class DouYuSpider():
    def __init__(self, start_url):
        self.start_url = start_url
        self.driver = webdriver.Chrome()
        
    
    def html_parse(self):
        """页面数据提取"""
        room_list = self.driver.find_elements_by_xpath('.//div[@id="live-list-content"]/ul/li')
        info = {}
        info_list = []
        for room in room_list:
            info['title'] = room.find_element_by_xpath('.//h3[@class="ellipsis"]').text
            info['name'] = room.find_element_by_xpath('.//div[@class="mes"]/p/span[@class="dy-name ellipsis fl"]').text
            info['num'] = room.find_element_by_xpath('.//div[@class="mes"]/p/span[@class="dy-num fr"]').text
            info['type'] = room.find_element_by_xpath('.//span[@class="tag ellipsis"]').text
            info['img_src'] = room.find_element_by_xpath('.//span[@class="imgbox"]/img').get_attribute("data-original")
            info['room_url'] =room.find_element_by_xpath('./a[@class="play-list-link"]').get_attribute("href")
            info_list.append(info)
            print(info)
        next_page_url = self.driver.find_elements_by_class_name("shark-pager-next")
        next_page_url = next_page_url[0] if len(next_page_url) > 0 else None
        return info_list,next_page_url
        
    def save_data(self, info_list):
        print("------ 保存成功！--------")
        
    def run(self):
        # 1.start_url
        # 2.发起请求
        try:
            self.driver.get(self.start_url)
            # 3.提取页面数据
            info_list,next_url = self.html_parse()
            self.save_data(info_list)
            i = 2
            while next_url is not None:
                print("正在爬取第 %i 页...."%i)
                next_url.click()
                time.sleep(1)
                info_list,next_url = self.html_parse()
                self.save_data(info_list)
                i += 1
        except:
            print("发现一个bug!")
            self.close()
        
    def close(self):
        self.driver.quit()
        print("%s爬虫结束！！！！%s" %("*"*20,"*"*20))
        
if __name__ == "__main__":
    start_url = "https://www.douyu.com/directory/all"
    douyu = DouYuSpider(start_url)
    douyu.run()
    douyu.close()
    
```
